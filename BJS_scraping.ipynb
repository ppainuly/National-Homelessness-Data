# Part 2
# Reading Homelessness Data from HUD Exchange
### Source : Excel Metadata

```python
# Read Data source
dfs = pd.read_excel('data/homelessness_data.xlsx')
```

```python
dfs.head()
```

```python
# Shape of the data set
dfs.shape
```

```python
# Remove all columns with null
dfs = dfs.dropna(axis=1, how='all')
```

```python
# Remove all columns with 0's
dfs = dfs.loc[:, (dfs != 0).any(axis=0)]
```

```python
dfs.head()
```

```python
dfs.columns
```

```python
# Clean the number of columns needed
# Assign the needed columns
dfs = dfs[['Row #', 'CocState', 'CoC', 'Coc\ID', 'HudNum', 'Status', 'year',
       'Organization ID', 'Organization Name', 'Project ID', 'Project Name',
       'Project Type', 'Bed Type', 'Geo Code', 'Inventory Type',
       'beginsOperationsWithinYear', 'Target Population A',
       'Target Population B','Victim Service Provider', 'address1', 'address2', 'city', 'state',
       'zip', 'Beds HH w/ Children', 'Units HH w/ Children',
       'HMIS Beds HH w/ Children', 'Veteran Beds HH w/ Children',
       'Youth Beds HH w/ Children', 'CH Beds HH w/ Children',
       'Beds HH w/o Children', 'HMIS Beds HH w/o Children',
       'Veteran Beds HH w/o Children', 'Youth Beds HH w/o Children',
       'CH Beds HH w/o Children', 'Beds HH w/ only Children',
       'HMIS Beds HH w/ only Children', 'CH Beds HH w only Children',
       'Year-Round Beds', 'DV Beds', 'Total Seasonal Beds',
       'Seasonal Beds Available in HMIS', 'Availability Start Date',
       'Availability End Date', 'O/V Beds', 'HMIS O/V Beds', 'PIT Count',
       'Total Beds', 'Updated On', 'questionUsesDescriptorElements',
       'questionDesiresToUseDescriptorElements', 'notes']]
```

```python
dfs.head()
```

```python
# Clean NaT Types in the dataframe
dfs[['Availability Start Date']] = dfs[['Availability Start Date']].astype(object).where(dfs[['Availability Start Date']].notnull(), None)
dfs[['Availability End Date']] = dfs[['Availability End Date']].astype(object).where(dfs[['Availability End Date']].notnull(), None)
```

```python
dfs.head(10)
```

```python
# Export Transformed Data to csv
dfs.to_csv('data/homelessness.csv',index=False)
```
